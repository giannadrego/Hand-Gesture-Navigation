```python
import os
import cv2
import time
import uuid

#Keep changing the number as per your set HandGestureDataset1, HandGestureDataset2 ... HandGestureDataset5
# 1) Any good lighting condition with face , 2) sitting opposite the light source (for ex back against window) with face , 3) Change the backgroound with/without face, 4) Change Tshirt color with/without face, 5) 1 set with only hands with no objects just a wall in background
IMAGE_PATH='HandGestureDataset6'

#All gestures are recorde using right hand 
labels=['fingerup','fingerdown','thumbsup','thumbsdown','fingerleft','fingerright','ok','closedfist','peace sign','rockon sign','L']
#labels=['fingerup','rockon sign','L']
number_of_images=20

for label in labels:
    img_path = os.path.join(IMAGE_PATH, label)
    os.makedirs(img_path)
    cap=cv2.VideoCapture(0)
    print('Collecting images for {}'.format(label))
    
    time.sleep(5)
    for imgnum in range(number_of_images):
        ret,frame=cap.read()
        cv2.putText(frame, f"Recording: {label} Number:{imgnum}", (10, 30),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)
        imagename=os.path.join(IMAGE_PATH,label,label+'.'+'{}.jpg'.format(str(uuid.uuid1())))
        cv2.imwrite(imagename,frame)
        cv2.imshow('frame',frame)
        time.sleep(2)
        
        if cv2.waitKey(1) & 0xFF==ord('q'):
            break
    cap.release()
```

```python
import os
import shutil
import zipfile

def extract_zips_to_common_folder(parent_folder, common_folder):
    # Create the common folder if it doesn't exist
    os.makedirs(common_folder, exist_ok=True)

    # Iterate through each subfolder in the parent folder
    for gesture_folder in os.listdir(parent_folder):
        gesture_folder_path = os.path.join(parent_folder, gesture_folder)
        # Check if the subfolder is a directory
        if os.path.isdir(gesture_folder_path):
            print("Extracting zip files from folder:", gesture_folder)
            # Iterate through each file in the subfolder
            for filename in os.listdir(gesture_folder_path):
                # Check if the file is a zip file
                if filename.endswith(".zip"):
                    # Open the zip file
                    zip_file_path = os.path.join(gesture_folder_path, filename)
                    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
                        # Extract all contents to the common folder
                        zip_ref.extractall(common_folder)

                    print("Extracted contents from:", zip_file_path)

parent_folder = "/Users/aseemmishra/Desktop/ML Project/Hand-Gesture-Navigation/Gianna/GiannaLabels"
common_folder = "/Users/aseemmishra/Desktop/ML Project/Hand-Gesture-Navigation/Gianna/Labels"

extract_zips_to_common_folder(parent_folder, common_folder)
```

```python
import cv2
import os

def crop_images_in_folders(parent_folder, target_resolution):
    # Iterate through each subfolder in the parent folder
    for gesture_folder in os.listdir(parent_folder):
        gesture_folder_path = os.path.join(parent_folder, gesture_folder)
        # Check if the subfolder is a directory
        if os.path.isdir(gesture_folder_path):
            print("Processing images in folder:", gesture_folder)
            # Iterate through each file in the subfolder
            for filename in os.listdir(gesture_folder_path):
                # Check if the file is an image
                if filename.endswith(".jpg") or filename.endswith(".png") or filename.endswith(".jpeg"):
                    # Read the image
                    image_path = os.path.join(gesture_folder_path, filename)
                    img = cv2.imread(image_path)

                    # Calculate how much to crop from the top
                    top_crop = img.shape[0] - target_resolution[1]

                    # Crop the image to the target resolution starting from the top
                    cropped_img = img[top_crop:, :target_resolution[0]]

                    # Save the cropped image
                    cropped_image_path = os.path.join(gesture_folder_path, "cropped_" + filename)
                    cv2.imwrite(cropped_image_path, cropped_img)

                    print("Cropped image saved:", cropped_image_path)

parent_folder = "HandGestureDataset6"
target_resolution = (1280, 680)  # Specify the desired resolution (width, height) in pixels

crop_images_in_folders(parent_folder, target_resolution)
```

```python
import os
import shutil

def move_cropped_images_to_train_folder(parent_folder, train_folder_path):
    # Create the "train" folder if it doesn't exist
    os.makedirs(train_folder_path, exist_ok=True)

    # Iterate through each subfolder in the parent folder
    for gesture_folder in os.listdir(parent_folder):
        gesture_folder_path = os.path.join(parent_folder, gesture_folder)
        # Check if the subfolder is a directory
        if os.path.isdir(gesture_folder_path):
            print("Moving cropped images from folder:", gesture_folder)
            # Iterate through each file in the subfolder
            for filename in os.listdir(gesture_folder_path):
                # Check if the file is a cropped image
                if not filename.startswith("."):
                    # Move the cropped image to the "train" folder
                    cropped_image_path = os.path.join(gesture_folder_path, filename)
                    shutil.move(cropped_image_path, train_folder_path)

                    print("Moved cropped image:", cropped_image_path)

parent_folder = "/Users/aseemmishra/Desktop/ML Project/Hand-Gesture-Navigation/Gianna/HandGestureDataset4"
train_folder_path = "/Users/aseemmishra/Desktop/ML Project/Hand-Gesture-Navigation/Gianna/images"

move_cropped_images_to_train_folder(parent_folder, train_folder_path)
```
